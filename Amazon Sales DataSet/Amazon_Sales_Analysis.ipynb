```python
# Check whether columns have missing values
for col in amazon_df.columns:
    missing_value = amazon_df[col].isna().sum()
    if missing_value != 0:
        print(missing_value, 'values is/are missing in', col)
        print('Missing percentage of column', col, ' =', missing_value * 100 / len(amazon_df))
```
